```python
sq_price = []
sq_price2 = []
sq_price3 = []
for i in calc_sq_data:
    try:

        sq_raw = re.findall('\d+', str(i))

    except:
        sq_raw = 1
    sq_price.append(sq_raw)

for b in sq_price:
    try: 
        s = min(b)
    except:
        s = 1
    sq_price2.append(s)

for d in sq_price2:
    s = int(d)
    sq_price3.append(s)


sq_pr = []
sq_pr2 = []

for i in calc_price_data:
    
    t = str(i)
    r = t.replace(' ','')

    sq_pr.append(r)


for i in sq_pr:
    try:
        d = int(i)
    except:
        d = 1
    sq_pr2.append(d)


print(sq_pr2)

per_sq = [round(i,1) / round(j,1) for i, j in zip(sq_pr2, sq_price3)]

print(per_sq)

final_per_sq = []

for i in per_sq:
    d = round(i, 0)
    if d == 1:
        d = None
    
    final_per_sq.append(d)

```
